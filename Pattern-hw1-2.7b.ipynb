```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial.distance import mahalanobis

# Covariance matrix
Sigma = np.array([[1.2, 0.4], [0.4, 1.8]])
Sigma_inv = np.linalg.inv(Sigma)

# Mean
mu = np.array([2.1, 1.9])

# Create a grid of points
x, y = np.meshgrid(np.linspace(0, 4, 100), np.linspace(0, 4, 100))
pos = np.dstack((x, y))

# Mahalanobis distance function
def mahalanobis_distance(pos, mu, Sigma_inv):
    diff = pos - mu
    return np.sqrt(np.einsum('...i,ij,...j', diff, Sigma_inv, diff))

# Compute Mahalanobis distances
d = mahalanobis_distance(pos, mu, Sigma_inv)

# Plot
plt.contour(x, y, d, levels=[1, 2, 3], colors='r')
plt.scatter(mu[0], mu[1], color='blue')  # Center point
plt.title('Equal Mahalanobis Distance Curves')
plt.xlabel('x')
plt.ylabel('y')
plt.axis('equal')
plt.show()
```
